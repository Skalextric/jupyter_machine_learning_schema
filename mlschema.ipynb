```python
import pandas as pd

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data",
    names=names)
df.shape
```

```python
% matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import itertools
sns.set(color_codes=True)
sns.set_style("white")
plt.rcParams["figure.figsize"] = [12, 9]
#pp.rcParams["xtick.color"] = 'white'
#pp.rcParams["ytick.color"] = 'white'
palette = itertools.cycle(sns.color_palette())
```

```python
f, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.ravel()
for x, y in enumerate(df.columns):
    sns.distplot(df[y], color=next(palette), ax=axes[x])
```

```python
f, axes = plt.subplots(2, 4, figsize=(15, 15))
axes = axes.ravel()
for x, y in enumerate(df.drop('class', axis=1).columns):
    ax = sns.boxplot(df[y], color=next(palette), ax=axes[x], orient='vertical', linewidth=3)
    ax.set_ylabel('')
    ax.set_title(y)
f.tight_layout()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

array = df.values
X = array[:, 0:8]
Y = array[:, 8]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
rescaledX[0:5, :]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
array = df.values
X = array[:, 0:8]
Y = array[:, 8]
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=test_size, random_state=seed)
model = LogisticRegression().fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print('Accuracy: {:.3f}'.format((result * 100))) 
```
